###Converting releasing date to date time

```python
df['release_date'] = pd.to_datetime(df['release_date'])
#Dropping some unnecessary columns 
```

###Converting the Duration to seconds

```python
df["duration"]=df["duration_ms"].apply(lambda x:round(x/1000))
df_songs = df.copy()
df_songs.drop(['duration_ms','explicit','time_signature'], inplace = True, axis = 1)
df_songs
```

##Popularity

###10 most least songs

```python
sorted_df=df_songs.sort_values("popularity", ascending=True)
sorted_df = sorted_df[['name','popularity','artists','release_date']]
sorted_df = sorted_df.set_index('name')
sorted_df.head(10)
```

###Most Popular Songs

```python
most_popular=df_songs.query("popularity>90",inplace=False).sort_values("popularity",ascending=False)
most_popular = most_popular[['name','popularity','artists','release_date']]
most_popular = most_popular.set_index('name')
most_popular
```

#Data Visulaization

##Correlation heatmap between variable

```python
correlation_df = df_songs.drop(['mode','key'], axis = 1).corr(method = 'pearson')
plt.figure(figsize = (15,8))
sns.diverging_palette(230, 20, as_cmap=True)
hmp = sns.heatmap(correlation_df, annot = True, fmt = '.1g', center =1, linewidths=1, linecolor = 'white')
hmp.set_title ("Correlation heatmap between variable")
hmp.set_xticklabels(hmp.get_xticklabels(),rotation=90)
```

###Taking sample dataset to plot regression plots

```python
sample_df = df_songs.sample(int(0.010*len(df_songs)))
print(len(sample_df))
```

```python
plt.figure(figsize=(20,13))
sns.regplot(data=sample_df,y="loudness",x="energy",scatter_kws={"color": "green"}, line_kws={"color": "red"}).set(title="loundness vs energy")
```

```python
plt.figure(figsize=(20,13))
sns.regplot(data=sample_df,y="popularity",x="tempo",scatter_kws={"color": "blue"}, line_kws={"color": "orange"}).set(title="popularity vs tempo")
```
